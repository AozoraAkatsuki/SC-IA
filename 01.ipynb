```python
trn_path = Path(r'train_jpg/train_jpg')
csv_path = Path('meta/meta')
```

```python
fn = get_image_files(trn_path)
len(fn)
labels = pd.read_feather(csv_path + '/labels.fth')
labels.head()
```

```python
labels = labels.iloc[:,0:2]
labels.head()
# for x in range(194082):
#     labels['ID'][x] += '.jpg'
labels['ID'] += '.jpg'
```

```python
fn=np.array(fn)
fn_df=pd.DataFrame(fn, columns=['ID'])
```

```python
fn_df.head()
fn_df.shape
```

```python
fn_df['ID'] = fn_df['ID'].apply(lambda x: str(x)[20:])
```

```python
fn_df['ID'].head()
```

```python
fn_df.shape
```

```python
labels.shape
```

```python
new=pd.merge(labels, fn_df, how='inner', on=['ID'])
```

```python
new.head()
```

```python
dls = ImageDataLoaders.from_df(new, trn_path, bs = 256, seed = 42)
```

```python
dls.bs
```

```python
dls.show_batch()
```

```python
learn = cnn_learner(dls, resnet18, pretrained=False, metrics=[error_rate, accuracy])
```

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(1, lr_max = 3e-3)
```
